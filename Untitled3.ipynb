```python
import json

# Charger le fichier GeoJSON
with open("gadm41_CMR_3.json", "r", encoding="utf-8") as f:
    gadm_data = json.load(f)

# Filtrer les arrondissements de Douala
douala_arrondissements = {
    "type": "FeatureCollection",
    "features": [
        feature for feature in gadm_data["features"]
        if feature["properties"]["NAME_1"] == "Littoral" and feature["properties"]["NAME_2"] == "Wouri"
    ]
}

# Sauvegarder le nouveau fichier GeoJSON
with open("douala_arrondissements.json", "w", encoding="utf-8") as f:
    json.dump(douala_arrondissements, f, indent=4)

print("✅ Extraction terminée !")
```

```python
import json
import plotly.express as px

# Charger le fichier JSON des arrondissements
with open("douala_arrondissements.json", "r", encoding="utf-8") as f:
    json_data = json.load(f)

# Vérifier la structure du fichier
if "features" in json_data:
    print("✅ Le fichier contient des features, il peut être utilisé comme GeoJSON.")

    # Créer la carte avec un bon centrage
    fig = px.choropleth_mapbox(
        geojson=json_data,
        locations=[feature["properties"]["NAME_3"] for feature in json_data["features"]],
        featureidkey="properties.NAME_3",
        color_discrete_sequence=["lightgray"],  # Couleur neutre
        mapbox_style="open-street-map",
        center={"lat": 4.06, "lon": 9.71},  # Centrage ajusté pour inclure Douala 6
        zoom=9# Zoom plus grand pour bien voir chaque arrondissement
    )

    # Ajouter les contours des arrondissements
    fig.update_traces(marker_line_width=2, marker_line_color="black")

    fig.show()
else:
    print("⚠️ Le fichier JSON ne contient pas de 'features'. Vérifiez son format.")
```

```python
import json
import pandas as pd
import plotly.express as px

# Charger le fichier JSON des arrondissements
with open("douala_arrondissements.json", "r", encoding="utf-8") as f:
    geojson_data = json.load(f)

# Vérifier si le fichier contient des "features"
if "features" not in geojson_data:
    raise ValueError("⚠️ Le fichier JSON ne contient pas de 'features'. Vérifiez son format.")

# Créer un DataFrame vide en attendant les données réelles
df = pd.DataFrame({
    "arrondissement": [feat["properties"]["NAME_3"] for feat in geojson_data["features"]],
    "valeur": [0] * len(geojson_data["features"])  # Placeholder pour la couleur
})

# Générer la carte des arrondissements de Douala
fig = px.choropleth_mapbox(
    data_frame=df,
    geojson=geojson_data,
    featureidkey="properties.NAME_3",
    locations="arrondissement",
    color="valeur",  # Actuellement, une valeur par défaut (on mettra les vraies données plus tard)
    color_continuous_scale="viridis",  # Style similaire à la carte espagnole
    mapbox_style="open-street-map",
    center={"lat": 4.06, "lon": 9.71},  # Centrage sur Douala
    zoom=9  # Zoom adapté
)

# Ajouter les contours noirs autour des arrondissements
fig.update_traces(marker_line_width=2, marker_line_color="black")

# Afficher la carte
fig.show()
```

```python
import json
import pandas as pd
import plotly.express as px

# Charger le fichier GeoJSON des arrondissements
with open("douala_arrondissements.json", "r", encoding="utf-8") as f:
    geojson_data = json.load(f)

# Liste des arrondissements à afficher
arrondissements_douala = ["Douala 1", "Douala 2", "Douala 3", "Douala 4", "Douala 5"]

# Filtrer le DataFrame pour ne garder que Douala 1 à 5
df_ville = df.groupby('Arrondissement de résidence').size().reset_index(name='Nombre')
df_ville = df_ville[df_ville['Arrondissement de résidence'].isin(arrondissements_douala)]

# Vérifier s'il y a des données après filtrage
if df_ville.empty:
    raise ValueError("⚠️ Aucune donnée pour Douala 1 à 5 ! Vérifiez votre DataFrame.")

# Générer la carte des arrondissements de Douala avec les valeurs du DataFrame
fig = px.choropleth_mapbox(
    data_frame=df_ville,
    geojson=geojson_data,
    featureidkey="properties.NAME_3",
    locations="Arrondissement de résidence",
    color="Nombre",  # On colore la carte en fonction du nombre d'occurrences
    color_continuous_scale="viridis",  # Palette de couleurs adaptée
    mapbox_style="open-street-map",
    center={"lat": 4.06, "lon": 9.71},  # Centrage sur Douala
    zoom=10  # Zoom adapté
)

# Ajouter les contours noirs autour des arrondissements
fig.update_traces(marker_line_width=2, marker_line_color="black")

# Afficher la carte
fig.show()
```
